```python
df = pd.DataFrame(df)

plt.figure(figsize=(12, 8))
top_20_countries = df.sort_values(by='TotalCases', ascending=False).head(30)
sns.barplot(x='Country/Region', y='TotalCases', data=top_20_countries)
plt.title('Total Number of COVID-19 Cases in the Top 20 Countries/Regions')
plt.xlabel('Country/Region')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()

```

```python
plt.figure(figsize=(8, 8))
df.groupby('Continent')['TotalCases'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of COVID-19 Cases by Continent')
plt.ylabel('')
plt.show()

```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Population'], df['TotalCases'])
plt.title('Relationship between Population Size and Total Cases')
plt.xlabel('Population')
plt.ylabel('Total Cases')
plt.show()

```

```python
top_30_countries = df.sort_values(by='TotalCases', ascending=False).head(30)
plt.figure(figsize=(16, 12))
top_30_countries[['ActiveCases', 'TotalRecovered', 'TotalDeaths']].plot(kind='bar', stacked=True)
plt.title('Distribution of COVID-19 Cases by Status for the Top 30 Countries/Regions')
plt.xlabel('Country/Region')
plt.ylabel('Number of Cases')
plt.xticks(top_30_countries.index, top_30_countries['Country/Region'])
plt.legend(title='Case Status')
plt.show()


```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Continent', y='TotalCases', data=df)
plt.title('Distribution of Total COVID-19 Cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.show()

```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Tot Cases/1M pop'], bins=10, edgecolor='black')
plt.title('Distribution of COVID-19 Cases per 1 Million Population')
plt.xlabel('Total Cases per 1 Million Population')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
top_30_countries = df.sort_values(by='TotalDeaths', ascending=False).head(30)  # Sort by TotalCases and select the top 30 countries
plt.pie(top_30_countries['TotalDeaths'], labels=top_30_countries['Country/Region'], autopct='%1.1f%%')
plt.title('Distribution of Total Deaths by Country/Region (Top 30)')
plt.show()
```

```python
top_30_countries = df.sort_values(by='TotalCases', ascending=False).head(50)
plt.figure(figsize=(16, 6))
sns.lineplot(data=top_30_countries, x="Country/Region", y="TotalCases", label="Total Cases")
sns.lineplot(data=top_30_countries, x="Country/Region", y="TotalRecovered", label="Total Recovered")
plt.xticks(top_30_countries.index, top_30_countries['Country/Region'])
plt.title("Total Cases vs. Total Recovered by Country/Region")
plt.xlabel("Country/Region")
plt.ylabel("Number of Cases")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
